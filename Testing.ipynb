### Identify song from queries:

```python
_ = audioIdentification('query_recordings', 'fingerprints.txt', 'song_preds.txt')
```

### Evaluate:

```python
# Open the file for reading
with open('song_preds.txt', 'r') as file:
    count = 0
    for line in file:
        # Split the line by tabs
        columns = line.strip().split('\t')
        if len(columns) >= 4:  # Ensure there are at least 4 columns
            query = columns[0]
            preds = columns[1:]  # Other three columns

            # Compare the first column with the other three
            for pred in preds:
                if query.startswith("pop"):
                    if query[:9] == (pred[:-4]):
                        print(f"{query} matches {pred}")
                        count+=1
                if query.startswith("classical"):
                    if query[:15] == (pred[:-4]):
                        print(f"{query} matches {pred}")
                        count+=1


accuracy = (count / 182)
print('accuracy:', accuracy)
```
